## 2.Graph

### Histogram

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# TODO: จงแสดงกราฟ histogram ข้อมูล "อายุบ้าน" ของชุดข้อมูล boston_house กำหนดให้มี 20 แท่งข้อมูล
plt.figure(figsize=(10, 6))
sns.histplot(boston_df["AGE"], bins = 20, kde=True, color='skyblue')
plt.xlabel('Ages')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.grid(axis = 'y')
plt.show()
```

```python
# TODO: จงแสดงกราฟ histogram ข้อมูล อายุบ้าน ของชุดข้อมูล boston_house กำหนดให้มี 100 แท่งข้อมูล
plt.figure(figsize=(10, 6))
sns.histplot(boston_df["AGE"], bins = 100, kde=True, color='skyblue')
plt.xlabel('Ages')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.grid(axis = 'y')
plt.show()
```

### Histogram Analysis:

1. The histogram appears to have a peak in the central age range, suggesting that the most common age group in the dataset is somewhere in the middle.

2. The histogram indicateste a diverse range of ages in the dataset.

3. The distribution appears to be slightly right-skewed, meaning that there are more data points on the younger side (left) of the histogram, and fewer data points on the older side (right).

### Bar Plot

```python
# TODO: แสดงกราฟแท่ง เพื่อบอกจำนวนข้อมูล boston_house ที่อยู่ริม/ไม่ริม แม่น้ำ (CHAS)
plt.figure(figsize=(8, 6))

# Group the data by the 'CHAS' column and count the occurrences
grouped_data = boston_df.groupby('CHAS').size().reset_index(name='Count')

# Rename the 'CHAS' values for better readability
grouped_data['CHAS'] = grouped_data['CHAS'].map({0: 'Not Near River (CHAS=0)', 1: 'Near River (CHAS=1)'})

# Create a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='CHAS', y='Count', data=grouped_data, palette='Set1')
plt.title('Number of Houses Near and Not Near the River (CHAS)')
plt.xlabel('CHAS (0: Not Near River, 1: Near River)')
plt.ylabel('Count')

# Annotate the bars with the counts
for index, row in grouped_data.iterrows():
    plt.text(index, row['Count'] + 10, str(row['Count']), ha='center', va='bottom')

plt.show()
```

### Bar Plot Analysis:

1.The bar plot clearly displays the count of houses in two categories: those near the river (CHAS=1) and those not near the river (CHAS=0).

2.There is a significant difference in the number of houses near and not near the river. In this case, it appears that there are more houses not near the river.

### Box Plot

```python
# TODO: จากข้อมูล boston_house จงแสดง boxplot ของคอลัมน์ PRICE
plt.figure(figsize=(8, 6))
sns.boxplot(x=boston_df['PRICE'], color='skyblue')
plt.title('Box Plot of House Prices (PRICE)')
plt.xlabel('Price')
plt.show()
```

### Box Plot Analysis:

1.Median (Q2): the median house price is around $22k. This value splits the data into two equal halves, with 50% of the prices below this value and 50% above.

2.Interquartile Range (IQR): The box's height suggests that most house prices are concentrated within a range of approximately $18k (Q1) to $27k (Q3).

3.Whiskers: There appear to be some lower-priced houses that are outliers, as indicated by points below the lower whisker.

4.Outliers: There are a few houses with prices significantly lower than the majority of houses.

```python
# TODO: จากข้อมูล boston_house จงแสดง boxplot ของคอลัมน์ PRICE แบ่งตามอยู่ริม/ไม่ริม แม่น้ำ (CHAS)
# Remove rows with missing values in 'PRICE' column
data_cleaned = boston_df.dropna(subset=['PRICE'])

# Create a box plot of 'PRICE' categorized by 'CHAS'
data_cleaned.boxplot(column='PRICE', by='CHAS', figsize=(10, 6))
plt.title('Box Plot of House Prices by River Proximity (CHAS)')
plt.xlabel('River Proximity (CHAS)')
plt.ylabel('House Prices (PRICE)')
plt.show()
```

Analysis:

1.For houses not near the river (CHAS=0), the median price appears to be lower, as indicated by the centerline of the box. The majority of data points fall within the box's interquartile range (IQR), showing a relatively tighter price distribution.

2.For houses near the river (CHAS=1), the median price appears higher compared to houses not near the river. The price distribution also appears more spread out, with a larger IQR.

3.Whiskers: The range of prices for houses not near the river (CHAS=0) seems narrower, indicating less price variability compared to houses near the river (CHAS=1).

4.Outliers: There are some outlier data points with higher prices for houses not near the river. These outliers represent properties with significantly higher prices.

```python
# TODO: จากข้อมูล boston_house จงแสดงกราฟ histogram และ boxplot ของทุกคอลัมน์ที่เป็นตัวเลข

# Step 1: Select columns with numeric data types
numerical_columns = boston_df.select_dtypes(include=np.number)

# Step 2: Get the names of these numeric columns
numeric_column_names = numerical_columns.columns.tolist()

print("Numerical Variables:")
print(numeric_column_names)
```
