# bcrypt library

```python
# Line 1: Imports the bcrypt library, which contains functions for hashing passwords and checking password hashes securely.
import bcrypt
```

```python
# Line 3: Defines a function named hash_password that takes a parameter password (the plaintext password to be hashed).
# Line 4: Calls bcrypt.gensalt() to generate a new random "salt". A salt is random data that is combined with the password before it is hashed. This helps to protect against rainbow table attacks and ensures that even if two users have the same password, their hashes will be different.
# Line 5: Calls bcrypt.hashpw() to hash the password. password.encode('utf-8') converts the string password to bytes because hashpw requires byte input. The salt is used to make the hash more secure.
# Line 6: Returns the hashed password.

def hash_password(password):
    # Generate a salt
    salt = bcrypt.gensalt()
    # Hash the password
    hashed_password = bcrypt.hashpw(password.encode('utf-8'), salt)
    return hashed_password
```

```python
# Line 8: Defines a function named check_password with parameters input_password (the plaintext password to check) and hashed_password (the previously hashed password to check against).
# Line 9: Calls bcrypt.checkpw() to check if the plaintext input_password (once hashed) matches the hashed_password. It returns True if they match and False otherwise. Again, input_password.encode('utf-8') is used to convert the string to bytes.

def check_password(input_password, hashed_password):
    return bcrypt.checkpw(input_password.encode('utf-8'), hashed_password)
```

```python
# Line 12: Defines a string plain_password that contains a plaintext password to be hashed.

# Example usage
plain_password = "my_secure_password"
```

```python
# Line 14: Calls the hash_password function, passing plain_password as the argument. The resulting hashed password is assigned to hashed_password.

# Hashing the password
hashed_password = hash_password(plain_password)
```

```python
# Line 16: Calls the check_password function, passing a plaintext password ("my_secure_password") and the previously computed hashed_password as arguments. The result (a boolean indicating whether the passwords match) is assigned to check.
# Line 17-20: An if/else statement that prints "Password match!" if check is True and "Password do not match!" if check is False.

# Checking the password
check = check_password("my_secure_password", hashed_password)
if check:
    print("Password match!")
else:
    print("Password do not match!")
```
