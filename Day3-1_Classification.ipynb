```python
import pandas as pd
df = pd.read_csv('diabete.csv')
print(df.head(5))
```

```python
df.drop(['PatientID'], axis = 1, inplace = True)
```

```python
print(df.info())
```

```python
# check duplicate row
print(df.duplicated().sum())
```

```python
X = df.drop('HasDiabetes', axis = 1)
y = df['HasDiabetes']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=11)
```

```python
import pandas as pd
iris_df = pd.read_csv('iris.csv')
print(iris_df)
```

```python
# Check duplicate row
print(iris_df.duplicated().sum())
```

```python
# Handle duplicates
iris_df.drop_duplicates(inplace = True)
```

```python
iris_df = iris_df.drop("Id", axis = 1)
iris_X = iris_df.drop("Species", axis = 1)
iris_y = iris_df["Species"]
```

```python
d = {'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica': 2}
iris_y = iris_y.map(d)
```

```python
from sklearn.model_selection import train_test_split
iris_X_train, iris_X_test, iris_y_train, iris_y_test = train_test_split(iris_X, iris_y, test_size=0.2, random_state=11)
```

#1.1Naive Bayes

## Naive Bayes - Cross Validation

```python
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB

NB_model = GaussianNB()
cvs = cross_val_score(NB_model, X, y, cv=5)
print('cross val scores', cvs.round(3))
print('mean = ', cvs.mean().round(4) * 100)
print('std = ', (cvs.std() * 100).round(3))
```

## Naive Bayes - Train/Test Split

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB

NB_model = GaussianNB()
NB_model = NB_model.fit(X_train, y_train)

y_pred = NB_model.predict(X_test)
print('Accuracy Score: ', accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

## Analysis:

1.Cross-Validation Scores:
These scores [0.933 0.967 0.933 0.933 1.] signify the accuracy of the model for each of the five folds. The scores are significantly high, indicating strong performance of the model on each fold.

2.Mean Accuracy:
The mean accuracy 95.33 suggests that the model is, on average, able to correctly predict the class labels 95.33% of the time. This indicates a high level of performance of the model on the dataset.

3.Standard Deviation of Accuracy:
The standard deviation std = 2.667 is a measure of the amount of variation or dispersion in the accuracy scores. A lower value of standard deviation is usually desired, indicating consistent performance across different subsets of data. However, in this case, the standard deviation is relatively higher compared to the mean, signifying some level of inconsistency in the model's performance across different folds.
